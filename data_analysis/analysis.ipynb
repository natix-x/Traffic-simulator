### Statistics

```python
one_car_per_sec = [name for name in names if name.endswith('_1')]
tree_cars_per_sec = [name for name in names if name.endswith('_3')]

equal = [name for name in names if name.startswith('eq_')]

single = {'fixed': [], 'max_wait': [], 'most_cars': [], 'most_wait': []}
single['fixed'] = [name for name in names if name.startswith('sin') and len(name) == 6]
single['max_wait'] = [name for name in names if name.startswith('sin_max')]
single['most_cars'] = [name for name in names if name.startswith('sin_most_c')]
single['most_wait'] = [name for name in names if name.startswith('sin_most_w')]

opposite = {'fixed': [], 'max_wait': [], 'most_cars': [], 'most_wait': []}
opposite['fixed'] = [name for name in names if name.startswith('opp') and len(name) == 6]
opposite['max_wait'] = [name for name in names if name.startswith('opp_max')]
opposite['most_cars'] = [name for name in names if name.startswith('opp_most_c')]
opposite['most_wait'] = [name for name in names if name.startswith('opp_most_w')]

light_3 = [name for name in names if name.endswith('3_1') or name.endswith('3_3')]
light_5 = [name for name in names if name.endswith('5_1') or name.endswith('5_3')]
light_7 = [name for name in names if name.endswith('7_1') or name.endswith('7_3')]
```

```python
equal_mean = data[equal].mean(axis=1)
single_mean = data[sum(single.values(), [])].mean(axis=1)
opposite_mean = data[sum(opposite.values(), [])].mean(axis=1)

plt.figure()
plt.plot(equal_mean, color='red', label='equal')
plt.plot(single_mean, color='blue', label='single')
plt.plot(opposite_mean, color='green', label='opposite')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Vehicle count')
plt.title('Vehicle Count at Intersections by Intersection Type')
plt.show()
```

```python
def group_value(columns):
    return data[columns].iloc[-1].mean()

groups = ['equal', 'single', 'opposite']
intensities = ['one_car_per_sec', 'tree_cars_per_sec']

values = {group: [] for group in groups}

group_columns = {
    'equal': equal,
    'single': one_car_per_sec,
    'opposite': one_car_per_sec
}


group_columns = {
    'equal': equal,
    'single': single['fixed'],
    'opposite': opposite['fixed']
}

for group in groups:
    for intensity in intensities:
        if intensity == 'one_car_per_sec':
            cols = [col for col in group_columns[group] if col.endswith('_1')]
        else:
            cols = [col for col in group_columns[group] if col.endswith('_3')]

        val = data[cols].iloc[-1].mean()
        values[group].append(val)


x = np.arange(len(groups))
bar_width = 0.35

plt.figure(figsize=(10, 6))

plt.bar(x - bar_width/2, [values[g][0] for g in groups], width=bar_width, label='One car per sec')
plt.bar(x + bar_width/2, [values[g][1] for g in groups], width=bar_width, label='Three cars per sec')

plt.xticks(x, groups)
plt.xlabel('Type of intersection')
plt.ylabel('Number of cars (last timestep)')
plt.title('Number of cars passing intersections by type and traffic intensity')
plt.legend()

plt.tight_layout()
plt.show()
```
